```python
#Task-2 (Find Square or Cube)

def power(number , choice):
  if choice == "square":
    return number * number
  elif choice == "cube":
    return number * number * number

print(power(3 , "square"))
print(power(3 , "cube"))
```

```python
#Task -3 (Calculator)

def calculator(a , b, operator):
  if operator == "+":
    return a+b
  elif operator == "-":
    return a-b
  elif operator == "*":
    return a*b
  elif operator == "/":
    return a/b
  else:
    return "Invalid Operator"
print(calculator(10 , 3,"*"))
print(calculator(10 , 3,"+"))
print(calculator(10 , 3,"-"))
print(calculator(10 , 3,"/"))
print(calculator(10 , 3,"%"))
```

```python
#Sum of the lists with loop

def sum_of_list(numbers):
  total = 0

  for n in numbers:
    total +=n
  return total

print(sum_of_list([10,29,39]))
```

```python
#finding maximum number in the list
def finding_max(numbers):
  maximum = numbers[0]

  for n in numbers:
    if n > maximum:
      maximum = n
  return maximum
print(finding_max([1,2,3,4,5,6]))
```

```python
#removing duplicates in the list

def remove_dup(numbers):
  unique = []

  for n in numbers:
    if n not in unique:
      unique.append(n)
  return unique

print(remove_dup([1,2,2,2,3,4,5,5,5,5,6,6]))
```

```python
# Q8Ô∏è‚É£ Find Second Largest Number

# Function takes a list
# Returns second largest element.

def second_largest(lst):
    largest = lst[0]
    second_largest = -1   # assuming list has positive numbers

    for i in lst:
        if i > largest:
            second_largest = largest
            largest = i
            # print(largest, second_largest)
        elif i > second_largest and i != largest:
            second_largest = i

    return second_largest
print(second_largest([5, 1, 4, 3, 2]))
```

```python
# Q9Ô∏è‚É£ Word Frequency Function

# Function takes a sentence
# Returns a dictionary of word counts.

def frequency(sentence):
  freq = {}

  seprt = sentence.split()

  for i in seprt:
    if i in freq:
      freq[i] = freq[i] + 1
    else:
      freq[i] = 1


  return freq
print(frequency("ram is a good is a boy and a boy"))
```

```python
# Q9. Remove Duplicates from List

# Write a function that:
# Takes a list
# Returns a new list without duplicates

def remove_dup(lst):
    dup =[]

    for i in lst:
        if i not in dup:
            dup.append(i)
    return dup

remove_dup([1,1,2,2,2,3,4,4,5])
```

```python
def calculate_grade(*scores, **adjustments):
    # taking the scores from input times
    scores = []

    for i in range(3):
        score = int(input(f"enter the score {i+1} : "))
        scores.append(score)


#taking the adjustments
    adjustments = {}
    bonus_points = int(input("How many bonus point need to add : "))

    for bonus in range(bonus_points):
        bonus_keys= str(input("Enter the bonus name : "))
        bonus_values = int(input("Enter the bonus value : "))
        adjustments[bonus_keys] = bonus_values



    total =0
    for score in scores:
        total+=score
    average = total / len(scores)


    bonus = 0
    for bonus_values in adjustments.values():
        bonus += bonus_values


    final_grade = average + bonus
    return final_grade

result = calculate_grade(*scores, **adjustments)
print(f"Final Grade: {result:.2f}")
```
